## Calculating technical indicators

### Simple Moving Average (SMA)
The Simple Moving Average (SMA) is a popular technical analysis tool used in finance. It calculates the average price of a security over a specific time period by adding the closing prices and dividing by the number of periods. SMA helps smooth out short-term price fluctuations and identify the overall trend of a security. Traders and investors use SMA to make decisions such as determining entry and exit points for trades and confirming trend directions.  
In our specific case we will use a window of 20 days.

```python
df['SMA'] = df['Close'].rolling(window=20).mean()
```

### Exponential Moving Average (EMA)
EMA is a type of moving average that gives more weight to recent price data. Unlike the Simple Moving Average (SMA), which assigns equal weight to all data points, the EMA places greater importance on the most recent prices. 
The EMA reacts more quickly to price changes compared to the SMA, making it more responsive to short-term trends.

```python
df['EMA'] = df['Close'].ewm(span=20, adjust=False).mean()
```

### Relative Strength Index (RSI)
RSI is a momentum oscillator that measures the speed and change of price movements. It is used to assess the overbought or oversold conditions of a financial instrument. RSI values range from 0 to 100 and are typically plotted on a graph. A reading above 70 is considered overbought, indicating a potential reversal or correction, while a reading below 30 is considered oversold, indicating a potential upward price movement.

```python
delta = df['Close'].diff()
gain = delta.mask(delta < 0, 0)
loss = -delta.mask(delta > 0, 0)
average_gain = gain.rolling(window=20).mean()
average_loss = loss.rolling(window=20).mean()
relative_strength = average_gain / average_loss
df['RSI'] = 100 - (100 / (1 + relative_strength))
```

### Bollinger Bands
Bollinger Bands are a technical analysis tool consisting of three lines plotted on a price chart. The middle line is a moving average, typically with a window of 20 periods, and the upper and lower lines are positioned based on standard deviations. They help identify overbought and oversold levels and gauge market volatility. Traders use them to spot potential reversals and assess market conditions.

```python
rolling_mean = df['Close'].rolling(window=20).mean()
rolling_std = df['Close'].rolling(window=20).std()
df['Upper Band'] = rolling_mean + (2 * rolling_std)
df['Lower Band'] = rolling_mean - (2 * rolling_std)
```

We encounter a notable challenge as the scraped data for macroeconomic indicators is not in the same frequency as our daily stock price data. This discrepancy is expected since obtaining daily data for certain macroeconomic indicators can be challenging. To address this issue, we have decided to convert all the data into a daily frequency.

Specifically, the CPI/inflation and interest rate data are provided on a monthly basis, while the GDP data is available on a quarterly basis. To align these data with our daily stock price data, we will perform a conversion process. For instance, if we have 1Q 2023 GDP data and a stock price for 05-01-2023, we will assign the GDP value for 1Q 2023 to the corresponding date of 05-01-2023.

This conversion allows us to integrate the macroeconomic indicators into our analysis and align them with the daily stock price data, enabling a comprehensive evaluation of their potential impact on stock prices.

```python
# Convert real GDP data to daily frequency
real_gdp['date'] = pd.to_datetime(real_gdp['date'])
real_gdp.set_index('date', inplace=True)
real_gdp = real_gdp.resample('D').ffill().reset_index()

# Convert CPI/inflation data to daily frequency
inflation['date'] = pd.to_datetime(inflation['date'])
inflation.set_index('date', inplace=True)
inflation = inflation.resample('D').ffill().reset_index()

# Convert interest rate data to daily frequency
interest_rate['date'] = pd.to_datetime(interest_rate['date'])
interest_rate.set_index('date', inplace=True)
interest_rate = interest_rate.resample('D').ffill().reset_index()
```

Another challenge arises as the final dates of our price data extend further into the future compared to the available macroeconomic data. To address this issue, we have decided to forward fill the values of the macroeconomic data, assuming that they remain relatively constant for the purpose of our analysis. 

For example, we have data points until 01-05-2023 for inflation and interest rate, while our GDP data only extends until 31-03-2023. In this case, we will forward fill the macroeconomic data until 26-06-2023, assuming that there is no significant change between the available data and the missing data for Q2.

By performing the forward filling process, we aim to align the dates of our price data and macroeconomic data, enabling a more synchronized analysis. While this approach assumes the relative stability of the macroeconomic indicators during the gap period, it allows us to proceed with the analysis by considering the available information.

### Sliding window
When working with time series data, such as stock prices, using traditional train-test split methods may not capture the dynamics of future patterns and developments. This poses a problem, particularly in the case of stocks like Apple, which has experienced rapid growth in recent years. To address this issue, we can employ the sliding window approach.

The sliding window approach involves creating a moving window of data that moves incrementally through the time series. This allows us to include both past and recent observations in our training data, capturing the changing patterns and trends. By sliding the window forward in time, we can train the model on a series of overlapping windows, providing a more comprehensive representation of the data and allowing the model to learn from past and present information.

This approach helps to account for the evolving nature of the stock market and ensures that the model is trained on data that reflects the latest developments. By incorporating the sliding window technique, we can improve the model's ability to capture and predict future trends in the stock price accurately.

```python
# Define the number of previous time steps to use for predicting the next time step
n_steps = 30

# Prepare the training data
X_train = []
y_train = []
for i in range(n_steps, len(train_data)):
    X_train.append(train_data[i - n_steps:i, 0])
    y_train.append(train_data[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

# Reshape the training data to match the input shape of the LSTM model
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

### LSTM model architecture

```python
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
```

The use of LSTM layers is suitable for capturing temporal dependencies and patterns in time series data.  
The addition of dropout layers helps prevent overfitting by randomly dropping out a fraction of the neurons during training, reducing their reliance on specific input features and improving the model's generalization ability.
The final dense layer with a single unit is responsible for predicting the target variable, in this case, the closing price. It maps the features learned by the preceding LSTM layers to the predicted output.

### Compiling and training
We are using EarlyStopping if the validation loss plateus to save computational resources.

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
early_stopping = EarlyStopping(monitor='val_loss', patience=10)
# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

### Testing data prep
We need to match the training dataset prep we did earlier with the sliding window.
We are using a similar approach with the testing data.

```python
# Prepare the testing data
X_test = []
y_test = []
for i in range(n_steps, len(test_data)):
    X_test.append(test_data[i - n_steps:i, 0])
    y_test.append(test_data[i, 0])
X_test, y_test = np.array(X_test), np.array(y_test)

# Reshape the testing data to match the input shape of the LSTM model
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

### Predictions
Using the trained model we are predicting y_test through X_test.
We need to inverse transform the data to "unscale" it from our prior preprocessing.  
In the last line of code we create a variable called test_dates which we get from the original merged_data dataframe.
